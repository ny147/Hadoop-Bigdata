```python
words = lines.flatMap(lambda line : list(line))
```

```python
vol = ["a","e","i","o","u"]
```

```python
def count_vol(alphabet):
  alphabet_lower = alphabet.lower()
  if alphabet_lower in vol:
    return alphabet_lower
  
```

```python
vol_map = words.filter(count_vol)
```

```python
vol_map.collect()
```

```python
tuples = vol_map.map(lambda word : (word,1))
# for lines in tuples.collect():
#   print(lines)
```

```python
counts = tuples.reduceByKey(lambda a, b: (a + b))
```

```python
counts.collect()
```

```python
counts.coalesce(1).saveAsTextFile(path+'outputDir2')
```
